#### 测试 PRC 的突触模型

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from import_fun import syn_prc, generate_pulses
```

```python
np.random.seed(1) 
```

```python
N = 2
dt = 0.01
tau_peak = .5
syn = syn_prc(N=N, dt=dt, tau_peak=tau_peak)
syn.params_syn["tau_r"] = 0.5
syn.params_syn["tau_d"] = 2.
s_list = []
s1_list = []

n = 700
t_list = np.arange(0, n*dt, 0.01)
for i in range(n):
    t = i * dt
    syn(t, ts=[0.5, 1.])
    s_list.append(syn.vars_syn[1, 0])
    s1_list.append(syn.vars_syn[1, 1])
```

```python
fig, axs = plt.subplots(2, sharex="all", layout='constrained')
axs[0].plot(t_list, s_list)
axs[1].plot(t_list, s1_list)
plt.show()
```

 #### 测试 PRC 的脉冲

```python
dt = 0.01
s1_list = []
n = 700
t_list = np.arange(0, n*dt, 0.01)
for i in range(n):
    t = i * dt
    s = generate_pulses(t, ts=[0.5, 2.], pulse_widths=1.)
    s1_list.append(s)
```

```python
s1_list = np.array(s1_list).T
```

```python
fig, axs = plt.subplots(2, sharex="all", layout='constrained')
axs[0].plot(t_list, s1_list[0])
axs[1].plot(t_list, s1_list[1])
plt.show()
```
